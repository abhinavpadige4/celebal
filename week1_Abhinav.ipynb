- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # YOUR CODE HERE
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None
    

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

# np.linalg.eig returns eigenvalues and column-eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair — the residual should be numerically zero
for i, (lam, vec) in enumerate(zip(eigenvalues, eigenvectors.T)):
    residual = np.linalg.norm(A @ vec - lam * vec)
    print(f"  Eigenpair {i+1}: λ={lam:.2f}, residual ‖Av−λv‖ = {residual:.2e}")

# Ploting eigenvectors as arrows scaled by their eigenvalues
fig, ax = plt.subplots(figsize=(6, 6))
colors = ['tomato', 'steelblue']
for i, (lam, vec, col) in enumerate(zip(eigenvalues, eigenvectors.T, colors)):
    scaled = lam * vec   # scale direction by eigenvalue magnitude
    ax.quiver(0, 0, scaled[0], scaled[1],
              angles='xy', scale_units='xy', scale=1,
              color=col, width=0.012,
              label=f'λ{i+1}={lam:.1f}')
ax.set_xlim(-6, 6); ax.set_ylim(-6, 6)
ax.axhline(0, color='k', lw=0.6); ax.axvline(0, color='k', lw=0.6)
ax.set_title('Eigenvectors scaled by Eigenvalues')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.legend()
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

When a matrix transforms things, it usually twists and turns vectors. But for special vectors, called **eigenvectors**, the matrix only stretches or shrinks them; their direction doesn't change. Think of the **eigenvalue** as the 'stretch factor' – it tells you exactly how much the eigenvector gets longer or shorter, or even flips direction. So, an eigenvalue of 2 means the eigenvector gets twice as long, while an eigenvalue of -0.5 means it shrinks by half and points the other way. It's like these eigenvectors are the fundamental axes of change for the matrix, along which transformations are simple scaling operations.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)
X_reconstructed = U @ np.diag(S) @ Vt
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD and PCA Connection (4.4):**

In SVD, **Vt** (rows) are the *right singular vectors* – they are exactly the **principal component directions** (the directions of maximum variance in the feature space). **U** contains the coordinates of the original data on the projected directions (the *scores*), and **S** contains the singular values whose squares are proportional to the variance explained by each component. Specifically, if you mean-center the data matrix **X**, then `PCA = U * diag(S)` gives you the principal component scores, and the columns of **Vt** are the loadings. The first row of **Vt** (the largest singular value) captures the direction of greatest variance, and that is why a rank-1 SVD is the best rank-1 approximation of the original matrix in the Frobenius norm sense.

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # Build common bin edges from the full combined range
    breakpoints = np.linspace(
        min(expected.min(), actual.min()),
        max(expected.max(), actual.max()),
        bins + 1
    )
    eps = 1e-10

    # Count observations falling into each bin, then convert to proportions
    exp_counts = np.histogram(expected, bins=breakpoints)[0]
    act_counts = np.histogram(actual,   bins=breakpoints)[0]

    exp_pct = exp_counts / len(expected) + eps
    act_pct = act_counts / len(actual)   + eps

    psi = np.sum((act_pct - exp_pct) * np.log(act_pct / exp_pct))
    return psi


np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

fig, ax = plt.subplots()
ax.hist(train_dist, bins=40, alpha=0.55, color='steelblue',
        density=True, label='Train (μ=50, σ=10)')
ax.hist(drift_dist, bins=40, alpha=0.55, color='tomato',
        density=True, label='Production (μ=65, σ=12)')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The underlying relationship between input features and the target label changes over time — e.g., customer behaviour that used to signal churn no longer does, so the model's decision boundary becomes wrong even if feature distributions stay the same.
- **Covariate drift:** The distribution of the *input features* shifts relative to what the model was trained on, even if the conditional relationship P(y|X) is unchanged — e.g., a sudden influx of older customers when the model was trained mostly on young ones.
- **PSI < 0.1:** Distribution is stable — no significant change detected; model can be trusted.
- **PSI 0.1–0.2:** Minor shift — worth investigating; consider monitoring more closely or collecting fresh labelled data.
- **PSI > 0.2:** Major shift — the production distribution has diverged significantly from training; immediate model retraining or replacement is recommended.
- **Retraining trigger example:** Automate weekly PSI computation on the top-5 most important features; trigger a retraining pipeline whenever PSI exceeds 0.2 on any one of them.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red   / total
p_blue  = blue  / total
p_green = green / total

p_red_then_blue  = (red / total) * (blue / (total - 1))

p_blue_given_red = blue / (total - 1)

independent = abs(p_blue_given_red - p_blue) < 1e-9   # False

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# Law of total probability: P(free) = P(free|spam)*P(spam) + P(free|ham)*P(ham)
p_free = p_free_given_spam * p_spam + p_free_given_ham * p_ham

# Bayes' theorem: P(spam|free) = P(free|spam) * P(spam) / P(free)
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    """Returns P(Spam | word) via Bayes' theorem."""
    prior_ham = 1 - prior_spam
    # Marginalise over both classes to get P(word)
    p_word = p_word_given_spam * prior_spam + p_word_given_ham * prior_ham
    # Posterior = likelihood × prior / evidence
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|------------------|
| Prior | P(Spam) | Our belief before seeing any word — 30 % of emails are spam |
| Likelihood | P(word \| Spam) | How often spammers use the word "free" — 80 % of spam contains it |
| Evidence | P(word) | Overall probability that any email (spam or not) contains "free" — computed via total probability law |
| Posterior | P(Spam \| word) | Updated belief after observing "free" — the probability the email is spam given it contains "free" |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples   = 5000
sample_size = 30

# Vectorised: draw all samples at once with reshape, then take row-wise means
sample_means = population[
    np.random.randint(0, len(population), size=(n_samples, sample_size))
].mean(axis=1)

pop_mean = population.mean()
pop_std  = population.std()
# CLT tells us sample means have std = σ_population / √n
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot —
fig, ax = plt.subplots()
ax.hist(sample_means, bins=60, density=True,
        color='blue', alpha=0.6, label='Observed sample means')
xs = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(xs, stats.norm.pdf(xs, pop_mean, clt_std),
        color='tomato', lw=2.5, label=f'CLT normal N({pop_mean:.2f},{clt_std:.3f})')
ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> Even though the underlying population is heavily skewed exponential, the distribution of *sample means* approaches a bell curve as the sample size increases. This result is foundational to statistics and machine learning because it justifies using the normal distribution to construct confidence intervals, run t-tests, and compute p-values — all of which assume normality of the estimator, not of the raw data.  It also accounts for the general behavior of stochastic gradient descent with mini-batches: the gradient estimates are noisy but cancel out on average, giving the true gradient.
